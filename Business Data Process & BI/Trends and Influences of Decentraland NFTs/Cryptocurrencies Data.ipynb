# **LUNAR: API DOWN FOR THE MOMENT**

```python
import urllib.request
import ssl
import json
import time
import tweepy


ssl._create_default_https_context = ssl._create_unverified_context


 
api_key = "API_KEY_HERE"

# Allows adding as many coins as desired
coin_list = [
    "MANA"
]
coins = ','.join(coin_list)


map = [
    {"name":""},
    {"symbol": ""},
    {"price": " Price: "},
    {"percent_change_24h": " - 24 Hour Percent Change: "},
    {"market_cap": " Market Cap: "},
    {"volume_24h": " 24 Hour Volume: "},
    {"url_shares": " URL Shares: "},
    {"reddit_posts": " Reddit Posts: "},
    {"tweets": " Tweets: "},
    {"galaxy_score": " Galaxy Score: "},
    {"volatility": " Volatility: "},
    {"social_volume": " Social Volume: "},
    {"news": " News: "},
    {"close": " Close: "},
]

def final_render(asset_coin, value, key, asset):
    if key == 'symbol':
        asset_coin += " (" + asset[key] + ")"
    elif key == 'percent_change_24h':
        asset_coin += value + str(asset[key]) + "%"
    else:
        asset_coin += value + str(asset[key])
    return asset_coin


# Iterates over each of the fields from Lunar Crush, gets the value from Lunar Crush and renders it with the field name
def main():

    url = "https://api.lunarcrush.com/v2?data=assets&key=" + api_key + "&symbol=" + coins
    assets = json.loads(urllib.request.urlopen(url).read())

    for asset in assets['data']:
        asset_coin = ""
        for field in map:
            key = list(field.keys())[0]
            value = list(field.values())[0]
            asset_coin = final_render(asset_coin, value, key, asset)
        print(asset_coin)
        print(len(asset_coin))


    # Runs main() every 30 minutes
while True:
    main()
    time.sleep(1800)
```

Regarding Marketplace place data we will retrieve historical data from 2019-11-15 to 2021-11-15: two years.

```python
# Retrieve historical data for decentraland

# Call API
coin_mana = cg.get_coin_market_chart_range_by_id(id='decentraland',vs_currency='usd',from_timestamp='1573779661',to_timestamp='1636938061')

df_mana = pd.DataFrame()

# Loop through keys
for i in range(0,len(coin_mana.keys())):
  key = coin_mana[list(coin_mana.keys())[i]]

  #Create df
  df_key = pd.DataFrame(key, columns=['timestamp', str(list(coin_mana.keys())[i])])

  # Reformat Data 
  df_key['date'] = df_key['timestamp'].apply(lambda x: datetime.datetime.fromtimestamp(x / 1e3))
  df_key = df_key.iloc[:, [2,1]]

  # Display df_key
  df_mana['date'] = df_key['date']
  df_mana = df_mana.join(df_key.iloc[:,1], rsuffix=str(list(coin_mana.keys())[i]))
  
df_mana['id'] = 'decentraland'
df_mana
```

```python
# Retrieve historical data for bitcoin

# Call API
coin_btc = cg.get_coin_market_chart_range_by_id(id='bitcoin',vs_currency='usd',from_timestamp='1573779661',to_timestamp='1636938061')

df_btc = pd.DataFrame()

# Loop through keys
for i in range(0,len(coin_btc.keys())):
  key = coin_btc[list(coin_btc.keys())[i]]

  #Create df
  df_key = pd.DataFrame(key, columns=['timestamp', str(list(coin_btc.keys())[i])])

  # Reformat Data 
  df_key['date'] = df_key['timestamp'].apply(lambda x: datetime.datetime.fromtimestamp(x / 1e3))
  df_key = df_key.iloc[:, [2,1]]

  # Display df_key
  df_btc['date'] = df_key['date']
  df_btc = df_btc.join(df_key.iloc[:,1], rsuffix=str(list(coin_btc.keys())[i]))
  
df_btc['id'] = 'bitcoin'
df_btc
```

```python
# Retrieve historical data for ethereum

# Call API
coin_eth = cg.get_coin_market_chart_range_by_id(id='ethereum',vs_currency='usd',from_timestamp='1573779661',to_timestamp='1636938061')

df_eth = pd.DataFrame()

# Loop through keys
for i in range(0,len(coin_eth.keys())):
  key = coin_eth[list(coin_eth.keys())[i]]

  #Create df
  df_key = pd.DataFrame(key, columns=['timestamp', str(list(coin_eth.keys())[i])])

  # Reformat Data 
  df_key['date'] = df_key['timestamp'].apply(lambda x: datetime.datetime.fromtimestamp(x / 1e3))
  df_key = df_key.iloc[:, [2,1]]

  # Display df_key
  df_eth['date'] = df_key['date']
  df_eth = df_eth.join(df_key.iloc[:,1], rsuffix=str(list(coin_eth.keys())[i]))
  
df_eth['id'] = 'ethereum'
df_eth
```

```python
# Append tables
df_market = pd.DataFrame()
df_market = df_market.append([df_mana,df_btc,df_eth])
df_market.reset_index(inplace=True)

#Display final dataset
df_market
```

```python
# Export dataset
df_market.to_csv('/content/drive/MyDrive/BI Decentraland/Financial_Data.csv')
```
