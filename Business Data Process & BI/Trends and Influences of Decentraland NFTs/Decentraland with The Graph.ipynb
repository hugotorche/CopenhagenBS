```python
# Importing the necessary libraries

from gql import gql, Client
from gql.transport.requests import RequestsHTTPTransport
import requests
import json
import pandas as pd
import time
```

```python
# Pre-set gql and url

sample_transport=RequestsHTTPTransport(
    url='https://api.thegraph.com/subgraphs/name/decentraland/marketplace',
    verify=True,
    retries=5,
)
client = Client(
    transport=sample_transport
)

url='https://api.thegraph.com/subgraphs/name/decentraland/marketplace'
```

```python
# Get the last block on the Blockchain protocol (date: 11/19/2021)

query = ('''
    query {
    orders(first: 1 orderBy:updatedAt orderDirection:desc where: {status:sold}) {
    blockNumber
}
}
    ''')

r = requests.post(url, json={'query': query})
json_data = json.loads(r.text)
final_block = json_data['data']['orders'][0]['blockNumber']

print("The final Block is:", final_block)
```

```python
# check if GPU is running
import tensorflow as tf 
tf.test.gpu_device_name()
```

```python
# Loop through The Graph data with the block number

# instantiate the dataframe
df = pd.DataFrame() 

# pass the final block integer to our x variable for iterations 
x = int(final_block) 

# go through all blocks for the 1000 last blocks
while x > 13643021: 
    query = ('''
    query {
    orders(blockNumber: {number: ''' + str(x) + '''} first: 1000 orderBy: updatedAt, orderDirection: desc where: { status:sold category:wearable}) {
        id
        blockNumber
        buyer
        owner {
        id
        }
        category
        nftAddress
        price
        status
        createdAt
        updatedAt
        nft {
        id
        tokenId
        category
        contractAddress
        owner {
        id
        }
        name
        tokenURI
        }
        createdAt
        updatedAt
    }
    }
    ''')

    # Request data
    r = requests.post(url, json={'query': query})
    json_data = json.loads(r.text)

    # Create a temporary dataframe to later append to the final dataframe that compiles all 1000-row dataframes.
    df1 = pd.DataFrame()
    df1 = pd.DataFrame(json_data['data']['orders'])

    # unfold a subdict into a series of columns.
    # df1 = df1.join(df1['nft'].apply(pd.Series))

    # append your temp dataframe to your master dataset
    df = df.append(df1)

    # go to next block
    x = x-1 

df['updatedAt'] = df['updatedAt'].apply(lambda x: time.strftime('%Y-%m-%d', time.localtime(int(x))))
df.reset_index(inplace=True)
df = df.iloc[:, 1:]

print(df.shape)
print(df)
```

```python
# Download the csv file
df.to_csv('/content/drive/MyDrive/BI Decentraland/Decentraland_Wearable_Data.csv')
```

# **Data Processing**

## **Load Data**

```python
# Load datasets
ens = pd.read_csv('/content/drive/MyDrive/BI Decentraland/Decentraland_Ens_Data.csv')
estate = pd.read_csv('/content/drive/MyDrive/BI Decentraland/Decentraland_Estate_Data.csv')
parcel = pd.read_csv('/content/drive/MyDrive/BI Decentraland/Decentraland_Parcel_Data.csv')
wearable = pd.read_csv('/content/drive/MyDrive/BI Decentraland/Decentraland_Wearable_Data.csv')
```

## **ENS Processing**

```python
# Data pre-processing: date variables
ens['createdAt'] = ens['createdAt'].apply(lambda x: time.strftime('%Y-%m-%d', time.localtime(int(x))))
ens['createdAt'] = pd.to_datetime(ens['createdAt'])
ens['updatedAt'] = pd.to_datetime(ens['updatedAt'])
ens.info()
```

```python
int(ens['price'][0])
```

```python
# Data pre-processing: sales data
ens['price'] = pd.to_numeric(ens['price'], errors='coerce')
ens.info()
```

```python
# Data pre-processing: NFT column 

# Turn column into a dict 
ens['nft'] = ens['nft'].apply(eval)
ens['nft'].apply(type)

# Join Series to the DataFrame
ens = ens.join(ens['nft'].apply(pd.Series), rsuffix='_nft') 

# Do the same for Owner sub-column
ens = ens.join(ens['owner_nft'].apply(pd.Series), rsuffix='_owner_nft') 

# Drop the useless columns
ens = ens.iloc[:, 1:]
ens = ens.drop(['nft', 'owner_nft'], axis=1)

# Save DataFrame
ens.to_csv('/content/drive/MyDrive/BI Decentraland/Decentraland_Ens_Data_V2.csv')
```
